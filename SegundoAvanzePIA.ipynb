**Introdución General**

El consumo de alcohol se encuentra dentro de los diez principales factores de riesgo de muerte y discapacidad en todo el mundo. En el 2015, el consumo de alcohol provocó 44,700 muertes (6.5% del total) y 5.2% (1.5 millones) de los años de vida saludables perdidos (AVISA) en México. El grupo de edad con mayor carga atribuible a este factor fue el de 15 a 49 años.

Más allá de las consecuencias a la salud, el consumo prolongado de alcohol provoca pérdidas en el aspecto social y económico que son de vital importancia para la vida en sociedad. 

En México, el consumo de alcohol forma parte de la vida cotidiana y propone ser un tema de círculo vicioso que aumenta con seriedad, buscando ser lo más próximamente atendido, no solamente en su efecto sino en sus medidas preventivas. Específicamente, en estudiantes, representa una gran amenaza con las suficientes consecuencias negativas ante el rendimiento académico, influencia producto de un comportamiento socialmente aceptado, que afecta de forma directa la productividad personal y el desarrollo del grado y calidad estudiantil.

**Descripcion de los datos**

**Tipo de datos**: Binarios y numéricos

**Descripción de las columnas**: la base de datos se conforma por 30 columnas

* escuela - escuela del estudiante (binario: 'GP' - Gabriel Pereira o 'MS' - Mousinho da Silveira)
* sexo - sexo del estudiante (binario: "F" - mujer o "M" - hombre)
* edad - edad del estudiante (numérica: de 15 a 22 años)
dirección - tipo de domicilio del estudiante (binario: "U" - urbano o "R" - rural)
* famsize - tamaño de la familia (binario: "LE3" - menos o igual a 3 o "GT3" - más de 3)
* Pstatus - estado de convivencia de los padres (binario: "T" - convivencia o "A" - separación)
* Medu - educación de la madre (numérico: 0 - ninguna, 1 - educación primaria (4º grado), 2 - 5º a 9º grado, 3 - educación secundaria o 4 - educación superior)
* Fedu - educación del padre (numérico: 0 - ninguna, 1 - educación primaria (4º grado), 2 - 5º a 9º grado, 3 - educación secundaria o 4 - educación superior)
* Mjob - trabajo de la madre (nominal: "profesor", "sanitario", "servicios" civiles (por ejemplo, administrativo o policía), "en_casa" u "otro")
* Fjob - trabajo del padre (nominal: "profesor", "sanitario", "servicios" civiles (por ejemplo, administrativo o policía), "en_casa" u "otro")
* razón - razón para elegir este centro (nominal: cerca de 'casa', 'reputación' del centro, preferencia de 'curso' u 'otro')
* tutor - tutor del alumno (nominal: "madre", "padre" u "otro")
* traveltime - tiempo de viaje de casa a la escuela (numérico: 1 - <15 min., 2 - 15 a 30 min., 3 - 30 min. a 1 hora, o 4 - >1 hora)
* tiempo de estudio - tiempo de estudio semanal (numérico: 1 - <2 horas, 2 - 2 a 5 horas, 3 - 5 a 10 horas, o 4 - >10 horas)
* fracasos - número de fracasos de clase anteriores (numérico: n si 1<=n<3, sino 4)
* schoolsup - apoyo educativo adicional (binario: sí o no)
* famsup - apoyo educativo familiar (binario: sí o no)
* paid - clases extra pagadas dentro de la asignatura del curso (matemáticas o portugués) (binario: sí o no)
* actividades - actividades extraescolares (binario: sí o no)
* guardería - asistió a la guardería (binario: sí o no)
* superior - quiere cursar estudios superiores (binario: sí o no)
* internet - acceso a internet en casa (binario: sí o no)
* romántico - con una relación romántica (binario: sí o no)
* famrel - calidad de las relaciones familiares (numérico: de 1 - muy malo a 5 - excelente)
* freetime - tiempo libre después de la escuela (numérico: de 1 - muy bajo a 5 - muy alto)
* goout - salir con amigos (numérico: de 1 - muy bajo a 5 - muy alto)
* Dalc - consumo de alcohol en el día de trabajo (numérico: de 1 - muy bajo a 5 - muy alto)
* Walc - consumo de alcohol en fin de semana (numérico: de 1 - muy bajo a 5 - muy alto)
* salud - estado de salud actual (numérico: de 1 - muy malo a 5 - muy bueno)
* ausencias - número de ausencias escolares (numérico: de 0 a 93)
* G1: Calificación del primer periodo [0-20]
* G2: Calificación del segundo periodo [0-20]
* G3: Calificación del periodo final [0-20]
